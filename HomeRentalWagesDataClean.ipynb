### 1.2 Performing Some Data Cleaning Operations

```python
housingPrices.drop(housingPrices.iloc[:, 9:housingPrices.shape[1]-1], inplace=True, axis=1)
rentalPrices.drop(rentalPrices.iloc[:, 9:rentalPrices.shape[1]-1], inplace=True, axis=1)
housingPrices.drop(columns=['RegionType', 'StateName'], inplace=True)
rentalPrices.drop(columns=['RegionType', 'StateName'], inplace=True)
```

Splitting the wage dataset up by country, state, and county. (By Area type)
Piping that into separate files.

```python
wagesMultiIndex = wagesnSalaries.set_index(['AREA_TYPE'])
indices = wagesMultiIndex.index.unique()

for index in indices:
    wagesMultiIndex.xs(index).to_csv(f'./data/wages_level_{index}.csv', index=False)
```
